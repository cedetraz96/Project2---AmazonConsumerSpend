```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Linear Regression predict star ratings; score=1.1, r2=1.1, mse=1.79, and rmse=1.34
# Random Forest predict start ratings; score=1.167, r2=-0.244, mse=1.167, and rmse=1.08

# data to plot
n_groups = 4
linear = (1.1, 1.1, 1.79, 1.34)
random_forest = (1.167, -0.244, 1.167, 1.08)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, linear, bar_width,
alpha=opacity,  
color='b',
label='Linear Regression')

rects2 = plt.bar(index + bar_width, random_forest, bar_width,
alpha=opacity,
color='g',
label='Random Forest')

plt.xlabel('Scores')
plt.ylabel('Scores')
plt.title('Scores by Linear Regression and Random Forest')
plt.xticks(index + bar_width, ('Score', 'R2', 'MSE', 'RMSE'))
plt.legend()

plt.tight_layout()  

plt.show()
```
